```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./iris.csv', names=["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"])
```

```python

g = sns.pairplot(df, hue='Species', markers='+')
plt.show()
```

```python
g = sns.violinplot(y='Species', x='SepalLength', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='SepalWidth', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalLength', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalWidth', data=df, inner='quartile')
plt.show()
```

```python
X = df.drop(['Species'], axis=1)
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import Perceptron

ppn = Perceptron(max_iter=40, eta0=0.1, random_state=1)
ppn.fit(X_train, y_train)
```

```python
y_pred = ppn.predict(X_test)
```

```python
y_test
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```
